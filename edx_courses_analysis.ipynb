# edX online courses
edx is a free online course provider that has more 160 member univerties including the best univesities like harvard. It has courses nearly for every subject which overall consist of 3000+ courses with different subjects. 
More importantly the courses are categorized for master's and bachelor's degrees.  In addition there are courses that has more than 10,000 enrolled students.
The majority of edX courses are entirely free to access and most offer an optional paid verified certificate track with graded assignments and the opportunity to work towards a certificate for a fee that varies per course. The Verified track awards a certificate after you successfully pass the course. Verified certificates are a great way to showcase the skills and knowledge you've gained, and can help advance your professional or academic career.


```python
import os
import pandas as pd 
import numpy as np
import matplotlib
import matplotlib.pyplot as plot
```

```python
edx_df=pd.read_csv("edx_courses.csv")
```

Sample of the edX courses 

```python
sample=edx_df[['title','course_type','Level','language','course_length']].set_index('title')
sample.head(10)
```

```python
Harvard=edx_df['institution'].value_counts()['Harvard University']
MIT = edx_df['institution'].value_counts()['Massachusetts Institute of Technology']
Stanford=edx_df['institution'].value_counts()['Stanford University']
Queensland=edx_df['institution'].value_counts()['The University of Queensland']
Boston=edx_df['institution'].value_counts()['Boston University']
data={'institutions':['Harvard University','MIT','Stanford University','The University of Queensland','Boston University'],'Number of Courses':[Harvard,MIT,Stanford,Queensland,Boston]}
df=pd.DataFrame(data=data)
df.plot.bar(x='institutions', y='Number of Courses',rot=70,title= 'Courses from famous Universities in edX')
plot.show(block=True)
```

## Level check of courses in edX

```python
Introuductory= edx_df['Level'].value_counts()['Introductory']
Intermediate= edx_df['Level'].value_counts()['Intermediate']
Advanced=edx_df['Level'].value_counts()['Advanced']
Levels = ['Introuductory','Intermediate','Advanced']
NumberOfCourses= [Introuductory,Intermediate,Advanced]
my_circle = plot.Circle((0, 0), 0.7, color='white')
plot.pie(NumberOfCourses, labels=Levels, autopct='%1.1f%%',
        colors=['blue','lightblue','darkblue'])
p = plot.gcf()
p.gca().add_artist(my_circle)
plot.show()
```

More than half of edX courses are Introductory which shows that edX courses are a guidance for the every field or area a person wants to be professional in

```python
def quality(row):
    if row.institution == 'Harvard University':
        return "great"
    else:
        return "Good"   
```

```python
edx_df['Quality']= edx_df.apply(quality, axis=1)
edx_df[['institution', 'n_enrolled', 'Level','Quality']].head()
```

## Computer Sciene Courses of Best Universities

```python
HarvardCompCourses=((edx_df.institution== "Harvard University") & (edx_df.subject == "Computer Science")).sum()
standfordcompCourses= ((edx_df.subject == "Computer Science") & (edx_df.institution == "Stanford University")).sum()
MITcompCourses= ((edx_df.subject == "Computer Science") & (edx_df.institution == "Massachusetts Institute of Technology")).sum()
QueenslandcomCourses= ((edx_df.subject == "Computer Science") & (edx_df.institution == "The University of Queensland")).sum()
UBCcompCourses= ((edx_df.subject == "Computer Science") & (edx_df.institution == "University of British Columbia")).sum()
data={'institutions':['Harvard University','Stanford University','MIT','The University of Queensland','University of British Columbia'],'Number of Computer Science Courses':[HarvardCompCourses,standfordcompCourses,MITcompCourses,QueenslandcomCourses,UBCcompCourses]}
df=pd.DataFrame(data=data)
df.plot.bar(x='institutions', y='Number of Computer Science Courses',rot=70,title= 'Courses from famous Universities in edX')
plot.show(block=True)
```
